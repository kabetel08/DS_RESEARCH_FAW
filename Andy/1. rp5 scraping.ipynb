```python
def get_country_links(driver_path, countries):
    driver = webdriver.Chrome(driver_path)
    driver.get('https://rp5.ru/Weather_in_the_world')
    links = driver.find_elements_by_tag_name('a')
    link_texts = [link.get_attribute('text') for link in links]
    af_links = [link.get_attribute('href') for link in links if link.get_attribute('text') in countries]
    return af_links
```

```python
def get_archive_links(driver_path,country_links):
    archives = set()
    driver = webdriver.Chrome(driver_path)

    driver.get('https://rp5.ru/Weather_in_the_world')

    #loop over the countries
    for link in country_links:
        print('visiting',link)
        driver.get(link)
        region_links = [a.get_attribute('href') for a in driver.find_elements_by_class_name('href12')]

        #loop over regions in country
        for link2 in region_links:
            print('........visiting',link2)
            driver.get(link2)
            station_links = [a.get_attribute('href') for a in driver.find_elements_by_xpath("//div[@class='city_link']//a")]

            #loop over stations in region
            for link3 in station_links:
                print('................visiting',link3)
                driver.get(link3)

                #metar link
                try:
                    metar = driver.find_element_by_id('metar_link').get_attribute('href')
                    archives.add(metar)
                except sel_exc.NoSuchElementException:
                    print('no metar link for',link3)

                #archive link
                try:
                    arc = driver.find_element_by_id('archive_link').get_attribute('href')
                    archives.add(arc)
                except sel_exc.NoSuchElementException:
                    print('no archive link for',link3)
    return list(archives)
```

```python
def download_archives(archive_links, start_date, end_date, driver_path, dest_path):
    """
    Input: 
        archive_links: list of URLs to download from
        start_date: str format "DD.MM.YYYY"
        end_date: str format "DD.MM.YYYY"
        driver_path: str path to chromedriver
        dest_path: str path to destination folder
    Output: 
        Returns list of all archive links
        Downloads gzip files to dest_path
    """
    chrome_options = webdriver.ChromeOptions()
    prefs = {'download.default_directory' : dest_path}
    chrome_options.add_experimental_option('prefs', prefs)
    driver = webdriver.Chrome(driver_path, chrome_options=chrome_options)
    downloaded = []
    driver.implicitly_wait(100)
    #loop over archives
    for arch in archive_links:
        print('downloading data from',arch)
        driver.get(arch)
        if arch[-5:] == 'METAR':
            dwa = driver.find_element_by_id('tabMetarDLoad')
        else:
            dwa = driver.find_element_by_id('tabSynopDLoad')
        dwa.click()

        start_date = driver.find_element_by_id('calender_dload')
        end_date = driver.find_element_by_id('calender_dload2')
        start_date.clear()
        start_date.send_keys('01.01.2018')
        end_date.clear()
        end_date.send_keys(yest)

        csv = driver.find_elements_by_class_name('radio')
        csv[7].click()
        csv[9].click()

        to_gz = driver.find_element_by_class_name('download')
        to_gz.click()

        try:
            wait = WebDriverWait(driver,10)
            download = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[@id='f_result']//a")))
            download.click()
            downloaded.append(arch)
            print('download successful')
        except sel_exc.TimeoutException:
            print(arch,'timed out, trying again later')
            archive_links.append(arch)
```
