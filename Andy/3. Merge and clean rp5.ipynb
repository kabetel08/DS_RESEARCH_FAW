```python
files = os.listdir('./rp5 csvs')
files.remove('.DS_Store')
files.sort()
files
```

```python
len(files)
```

```python
dfs = []
path = './rp5 csvs/'
filenames = [path+file for file in files]
```

## Checking for HTML tags

#### There are some HTML tags which causes pandas to error when trying to read the csv; they are removed below

```python
for i,filename in enumerate(filenames):
    
    if i%10==0:
        print(i,'of',len(files),'(',filename,')')
        
    # Read in the file
    with open(filename, 'r') as file:
        filedata = file.read()

    # Replace the target string
    newdata = re.sub('<.*?>', ' ',filedata)
    
    if newdata is not filedata:
        print('html found in',filename)
    
        # Write the file out again
        with open(filename, 'w') as file:
            file.write(newdata)
```

## Reading CSVs

```python
loc_names = []
regions = []
countries = []
for f in filenames:
    loc_name = f[len(f)-f[::-1].index('/'):len(f)-f[::-1].index('.')-1]
    loc_names = loc_names + [loc_name]
    regions = regions + [loc_name[:loc_name.index(',')]]
    countries = countries + [loc_name[loc_name.index(',')+1:]]
```

```python
for filename in files:
    try:
        f = pd.read_csv(path+filename,skiprows=6,delimiter=';',index_col=False)
    except SyntaxError as syn:
        print(syn)
    col_names = dict(zip(f.columns,new_cols))
    f.rename(col_names,axis='columns',inplace=True)
    loc_name = filename[:len(filename)-filename[::-1].index('.')-1]
    region = loc_name[:loc_name.index(',')]
    country = loc_name[loc_name.index(',')+1:]
    f['region'] = np.repeat(region,len(f))
    f['country'] = np.repeat(country,len(f))
    dfs = dfs + [f]
```

```python
all_rp5 = pd.concat(dfs,axis=0,sort=False)
```

```python
out = open('all_rp5','wb')
pickle.dump(all_rp5,out)
out.close()
```
