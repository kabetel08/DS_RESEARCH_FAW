# Expanding gzip files

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import re
import os
import codecs
import urllib
import gzip
import shutil
```

```python
path = './rp5-gzfiles'
 
files = os.listdir(path)
if '.DS_Store' in files:
    files.remove('.DS_Store')
```

```python
files
```

```python
len(files)
```

```python
for file_num,file in enumerate(files):
    if file_num < 10:
        new_file_name = '00'+str(file_num)+'.csv'
    elif file_num < 100:
        new_file_name = '0' +str(file_num)+'.csv'
    else:
        new_file_name = str(file_num)+'.csv'
    print('expanding',new_file_name)
    with gzip.open('./rp5/'+file, 'rb') as f_in:
            with open('./rp5 csvs/'+new_file_name, 'wb') as f_out:
                shutil.copyfileobj(f_in, f_out)
```

```python
path = './rp5 csvs/'
csv_files = os.listdir(path)
csv_files.remove('.DS_Store')
csvs = [path+csv for csv in csv_files]
for csv in csvs:
    with open(csv,errors= 'ignore') as f:
        line = f.readline()
    lst = line.split(sep=' ')
    start = 3
    end = lst.index([x for x in lst if x[-1:]==','][1])+1
    name = "".join(lst[start:end])[:-1].replace('/','.')
    print(name)
    os.rename(csv,path+name+'.csv')
```
